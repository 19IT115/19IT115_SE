```python
import pandas as pd
```

```python
data = pd.read_csv("/content/breast_cancer.csv")
data.head()
```

```python
data.info()
```

```python
# If we look at this data set what is a problem in this
# 1. In diagnosis datatype is object when ever we apply ML or DL there should be no data is
#    in alphabatical format or in string format it should be in numeric format
# So we will convert it into numeric form
# we will convert M to 1 and B to 0
# because M = positive
# And B = Negetive
```

```python
data.isnull().sum()
```

```python
# we will drop last column
data.columns
```

```python
data.drop(["Unnamed: 32"], axis = 1, inplace = True)
```

```python
data.drop(["id"], axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
# we will convert the value of diagnosis to 0 and 1
# here sklearn is scikit learn it is use for ML
from sklearn.preprocessing import  LabelEncoder # library for encoding
le = LabelEncoder()  # created the object
data['diagnosis'] = le.fit_transform(data['diagnosis'])
# Using map function
# data['diagnosis'] = data['diagnosis'].map({"B" = 0, "M" = 1})
```

```python
data.head()
```

```python
# Now we are going to divide the data into features and target
# diagnosis is target and others are features
x = data.iloc[:,1:] # to take all the column except diagnosis -> first one is row and second on is column
y = data.iloc[:,0] 
```

```python
x
```

```python
y
```

```python
y.value_counts()
```

```python
# now we will devide the data
# we have to devide it randomly because we can not take it in squence
from sklearn.model_selection import train_test_split  # train_test_split will take rows randomly
# if we devide x and y there will be 4 variable
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.80, random_state = 1)
```

```python
ytrain.value_counts()
```

```python
len(ytrain)
```

```python
# Neural Network
# for this we will use library Tensorflow/keras
```

```python
import tensorflow.keras as tf
```

```python
# Now we will create the blank model of neural network
model = tf.models.Sequential()  # Sequential is model of neural network this will creates blank model
# we can add any number of layer that we want to add
# Add the layer
model.add(tf.layers.Dense(16, input_dim = 30, activation = 'relu')) # 16 is no. of layer in hidden layer
                                            # input_dim is no. of neurons in input layer 
model.add(tf.layers.Dense(32, activation = 'relu'))  # next hidden layer
model.add(tf.layers.Dense(1, activation = 'sigmoid')) # Output layer   

# Compile the model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ["accuracy"]) # optimizer -> gradent desend optimizer and adam is function which will add that

```

```python
# Train the model
model.fit(xtrain, ytrain, epochs = 500)  # epochs is number of iteration
```

```python
# Test the Model  or we have to do predictions from the testing data
ypred = model.predict(xtest)  # This output is exponential value
ypred = ypred.round()
ypred
```

```python
# to print actuel value
ytest
```

```python
# to compare this 2
model.evaluate(xtest, ytest)
```

```python
# Confusion Matrix 
# by this we will able to know how many sample of 0 are wrongly predicted or correctly predicted
# Same of 1 also
# this will always a squre matrix
# size of matrix = no. of class (in this case 2X2)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, ypred)
cm
```
